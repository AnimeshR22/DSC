```python
import seaborn as sns
import matplotlib.pyplot as plt
```

A. Distribution Plot
A histogram and a KDE plot are good for visualizing the distribution of a single variable:

```python
# Set the plot size
plt.figure(figsize=(10,6))

# Plot the distribution using seaborn
sns.histplot(data['Exam_Score'], kde=True, bins=20)

# Adding labels and title
plt.title('Distribution of Exam Scores')
plt.xlabel('Exam Score')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

B. Box Plot to Identify Outliers
A box plot will help visualize outliers:

```python
# Boxplot to visualize outliers in the Exam Score
plt.figure(figsize=(8,6))
sns.boxplot(x=data['Exam_Score'])

# Add labels and title
plt.title('Box Plot of Exam Scores')
plt.xlabel('Exam Score')

# Show the plot
plt.show()
```

```python
# Calculate the summary statistics
mean_score = data['Exam_Score'].mean()
median_score = data['Exam_Score'].median()
std_dev = data['Exam_Score'].std()

print(f'Mean Exam Score: {mean_score}')
print(f'Median Exam Score: {median_score}')
print(f'Standard Deviation: {std_dev}')
```

```python
# Calculate the Q1 (25th percentile) and Q3 (75th percentile)
Q1 = data['Exam_Score'].quantile(0.25)
Q3 = data['Exam_Score'].quantile(0.75)
IQR = Q3 - Q1

# Define the bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the outliers
outliers = data[(data['Exam_Score'] < lower_bound) | (data['Exam_Score'] > upper_bound)]
print(f"Number of outliers: {len(outliers)}")
print(outliers)
```

```python
# Plot the distribution
plt.figure(figsize=(10,6))
sns.histplot(data['Exam_Score'], kde=True, bins=20)

# Calculate mean and median
mean = data['Exam_Score'].mean()
median = data['Exam_Score'].median()

# Plot the mean and median
plt.axvline(mean, color='r', linestyle='--', label=f'Mean: {mean:.2f}')
plt.axvline(median, color='g', linestyle='-', label=f'Median: {median:.2f}')

plt.title('Exam Score Distribution with Mean and Median')
plt.xlabel('Exam Score')
plt.ylabel('Frequency')
plt.legend()

plt.show()
```
